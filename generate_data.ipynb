```python
#2通道
def gen_graph_3(trace,arr):
  list_temp=trace.split(",")
  for i in range(0,len(list_temp)-3):
    arr[int(list_temp[i])][int(list_temp[i+3])]=arr[int(list_temp[i])][int(list_temp[i+3])]+1
```
